```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

import re
import nltk 
import nltk
#means that NLTK (Natural Language Toolkit) is trying to download the stopwords corpus (a list of common words like "the", "a", "an", etc.). 
from nltk.util import pr
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer("english")
from nltk.corpus import stopwords  
import string
stopword = set(stopwords.words("english"))

df = pd.read_csv(r"C:\Hate_speech_recog\twitter_data.csv") #Use Raw String Literal: You can tell Python to treat the string as raw (i.e., ignore escape sequences)by prefixing the string with r
print(df.head())
print("/n/n")
df['labels'] = df['class'].map({0:"Hate Speech Detected",1:"Offensive language detected",3:"No hate and offensive speech"})
print(df.head())
```

```python
df = df[['tweet','labels']]
df.head()
```

```python
def clean(text):
    text=str(text).lower()
    text=re.sub(r'\[.*?\]','',text)
    text=re.sub(r'https?://\S+|www\.S+','',text)
    text=re.sub(r'<.*?>+','',text)
    
    #In regular expressions, backslashes are often used for special characters (like \w, \S, and \d), 
   # but in a normal string, they are escape sequences in Python (e.g., \n for newline). To avoid confusion,
   # we use raw strings (r"..."), which interpret backslashes as literal characters.

    text = re.sub(r'[%s]' %re.escape(string.punctuation), '', text)
    text=re.sub(r'\n','',text)
    text=re.sub(r'\w*\d\w*','',text)
    text=[word for word in text.split(' ') if word not in stopword]
    text="".join(text)
    text=[stemmer.stem(word) for word in text.split(' ')]
    text= " ".join(text)
    return text
df["tweet"] = df["tweet"].apply(clean)
print(df.head())
```

```python
x= np.array(df["tweet"])
y= np.array(df["labels"])

# Fill NaN in 'labels' with a default label (e.g., "Unknown")
df["labels"] = df["labels"].fillna("No Hate and offensive speech")
df["tweet"]=df["tweet"].fillna("No tweets")
cv=CountVectorizer()
x=cv.fit_transform(x)
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state =42)

clf =  RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train,y_train)
```
